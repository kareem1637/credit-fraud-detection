# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/creditcard_cleaned.csv")
```

## Data Structure

### Shape

* Columns: 31
* Rows: 283726

```python
df.shape
```

### Datatypes

all features are numeric

```python
df.info()
```

### Balance

```python
is_fraud = df["Class"]
fraud = is_fraud.value_counts()[1]
not_fraud = is_fraud.value_counts()[0]
fraud_ratio = round(fraud / df.shape[0] * 100, 2)

print(f"Fraud Transactions: {fraud}")
print(f"Non Fraud Transactions: {not_fraud}")
print(f"Fraud Ratio: {fraud_ratio}%")
```

## Summary Statistics

```python
df.describe()
```

## Distributions

### V1:V28

```python
df.iloc[:, 1:29].hist(figsize=(25, 25), bins=30)
```

### Time

```python
fig, axes = plt.subplots(2, 1, figsize=(25, 10))

sns.histplot(df[df["Class"] == 1]["Time"], ax=axes[0], kde=True, color="red")
axes[0].set_title("Fraud Transactions")

sns.histplot(df[df["Class"] == 0]["Time"], ax=axes[1], kde=True, color="blue")
axes[1].set_title("Non Fraud Transactions")

plt.show()
```

### Amount

```python
fig, axes = plt.subplots(2, 1, figsize=(25, 12))

sns.histplot(df[df["Class"] == 1]["Amount"], ax=axes[0], kde=True, color="red")
axes[0].set_title("Fraud Transactions")

sns.histplot(df[df["Class"] == 0]["Amount"], ax=axes[1], kde=True, color="blue")
axes[1].set_title("Non Fraud Transactions")

plt.show()
```

## Boxplots

```python
# Create a figure with a 3x10 grid of subplots
fig, axes = plt.subplots(10, 3, figsize=(10, 30))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through the features and create a boxplot for each
for i, feature in enumerate(df.columns[:30]):
    sns.boxplot(x="Class", y=feature, data=df, ax=axes[i])

plt.tight_layout()
plt.show()
```

## Correlation Matrix

```python
plt.figure(figsize=(32, 32))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
```

V10, V12, V14 and V17 have strong negative correlations with the "Class" label.

V4 and V11 have strong positive correlations with the "Class" label.
